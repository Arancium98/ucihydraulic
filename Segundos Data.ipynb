```python
import pandas as pd
import plotly.express as px
import plotly.figure_factory as ff
import warnings
import numpy as np
warnings.filterwarnings('ignore')
```

```python
PS1 = pd.read_csv('/home/arancium/ucihydraulic/Data/PS1.txt', sep = '\t')
PS2 = pd.read_csv('/home/arancium/ucihydraulic/Data/PS2.txt', sep = '\t')
PS3 = pd.read_csv('/home/arancium/ucihydraulic/Data/PS3.txt', sep = '\t')
PS4 = pd.read_csv('/home/arancium/ucihydraulic/Data/PS4.txt', sep = '\t')
PS5 = pd.read_csv('/home/arancium/ucihydraulic/Data/PS5.txt', sep = '\t')
PS6 = pd.read_csv('/home/arancium/ucihydraulic/Data/PS6.txt', sep = '\t')
EPS1 = pd.read_csv('/home/arancium/ucihydraulic/Data/EPS1.txt', sep = '\t')
FS1 = pd.read_csv('/home/arancium/ucihydraulic/Data/FS1.txt', sep = '\t')
FS2 = pd.read_csv('/home/arancium/ucihydraulic/Data/FS2.txt', sep = '\t')
TS1 = pd.read_csv('/home/arancium/ucihydraulic/Data/TS1.txt', sep = '\t')
TS2 = pd.read_csv('/home/arancium/ucihydraulic/Data/TS2.txt', sep = '\t')
TS3 = pd.read_csv('/home/arancium/ucihydraulic/Data/TS3.txt', sep = '\t')
TS4 = pd.read_csv('/home/arancium/ucihydraulic/Data/TS4.txt', sep = '\t')
VS1 = pd.read_csv('/home/arancium/ucihydraulic/Data/VS1.txt', sep = '\t')
CE = pd.read_csv('/home/arancium/ucihydraulic/Data/CE.txt', sep = '\t')
CP = pd.read_csv('/home/arancium/ucihydraulic/Data/CP.txt', sep = '\t')
SE = pd.read_csv('/home/arancium/ucihydraulic/Data/SE.txt', sep = '\t')

```

```python
ReshapedPS1 = PS1.values.reshape(-1,100)
meansPS1 = ReshapedPS1.mean(axis = 1)
SPS1 = pd.DataFrame(meansPS1.reshape(PS1.shape[0], -1))


ReshapedPS2 = PS2.values.reshape(-1,100)
meansPS2 = ReshapedPS2.mean(axis = 1)
SPS2 = pd.DataFrame(meansPS2.reshape(PS2.shape[0], -1))

ReshapedPS3 = PS3.values.reshape(-1,100)
meansPS3 = ReshapedPS3.mean(axis = 1)
SPS3 = pd.DataFrame(meansPS3.reshape(PS3.shape[0], -1))

ReshapedPS4 = PS4.values.reshape(-1,100)
meansPS4 = ReshapedPS4.mean(axis = 1)
SPS4 = pd.DataFrame(meansPS4.reshape(PS4.shape[0], -1))

ReshapedPS5 = PS5.values.reshape(-1,100)
meansPS5 = ReshapedPS5.mean(axis = 1)
SPS5 = pd.DataFrame(meansPS5.reshape(PS5.shape[0], -1))

ReshapedPS6 = PS6.values.reshape(-1,100)
meansPS6 = ReshapedPS6.mean(axis = 1)
SPS6 = pd.DataFrame(meansPS6.reshape(PS6.shape[0], -1))

ReshapedEPS1 = EPS1.values.reshape(-1,100)
meansEPS1 = ReshapedEPS1.mean(axis = 1)
SEPS1 = pd.DataFrame(meansEPS1.reshape(EPS1.shape[0], -1))

ReshapedFS1 = FS1.values.reshape(-1,10)
meansFS1 = ReshapedFS1.mean(axis = 1)
SFS1 = pd.DataFrame(meansFS1.reshape(FS1.shape[0], -1))

ReshapedFS2 = FS2.values.reshape(-1,10)
meansFS2 = ReshapedFS2.mean(axis = 1)
SFS2 = pd.DataFrame(meansFS2.reshape(FS2.shape[0], -1))

```

```python
df_Segundos_PS1 = SPS1.stack().reset_index(drop=True)
df_Segundos_PS2 = SPS2.stack().reset_index(drop=True)
df_Segundos_PS3 = SPS3.stack().reset_index(drop=True)
df_Segundos_PS4 = SPS4.stack().reset_index(drop=True)
df_Segundos_PS5 = SPS5.stack().reset_index(drop=True)
df_Segundos_PS6 = SPS6.stack().reset_index(drop=True)
df_Segundos_EPS1 = SEPS1.stack().reset_index(drop=True)
df_Segundos_FS1 = SFS1.stack().reset_index(drop=True)
df_Segundos_FS2 = SFS2.stack().reset_index(drop=True)

```

```python
Df_TS1_Segundos = TS1.stack().reset_index(drop=True)
Df_TS2_Segundos = TS2.stack().reset_index(drop=True)
Df_TS3_Segundos = TS3.stack().reset_index(drop=True)
Df_TS4_Segundos = TS4.stack().reset_index(drop=True)
df_VS1_Segundos = VS1.stack().reset_index(drop=True)
df_CE_Segundos = CE.stack().reset_index(drop=True)
df_CP_Segundos = CP.stack().reset_index(drop=True)
df_SE_Segundos = SE.stack().reset_index(drop=True)
```

```python
df_SE_Segundos
```

```python
df_Segundos = pd.concat([df_Segundos_PS1,
                         df_Segundos_PS2,
                         df_Segundos_PS3,
                         df_Segundos_PS4,
                         df_Segundos_PS5,
                         df_Segundos_PS6,
                         df_Segundos_EPS1,
                         df_Segundos_FS1,
                         df_Segundos_FS2,
                         Df_TS1_Segundos,
                         Df_TS2_Segundos,
                         Df_TS3_Segundos,
                         Df_TS4_Segundos,
                         df_VS1_Segundos,
                         df_CE_Segundos,
                         df_CP_Segundos,
                         df_SE_Segundos
                         ], axis = 1)
```
