# Corrrelation

```python
# Create a mask to hide the upper triangular part (including the diagonal)
corr = data.corr()

mask = np.tril(np.ones_like(corr, dtype=bool))

# Apply the mask to the correlation matrix
masked_corr = np.where(mask, corr, np.nan)

fig = ff.create_annotated_heatmap(
    z=masked_corr.round(2),
    x=list(corr.index.values),
    y=list(corr.columns.values),       
    xgap=5, ygap=5,
    zmin=-1, zmax=1,
    colorscale='RdYlGn',
    colorbar_thickness=3,
    colorbar_ticklen=0,
    showscale=False,  # Set this to False to hide the color scale for NaN values
    annotation_text=masked_corr.round(2),  # Add annotation text with values
    font_colors=['black', 'white'],  # Specify font colors for positive and negative values
)

fig.update_layout(title_text='<b>Lower Triangular Correlation Matrix (cont. features)<b>',
                  title_x=0.5,
                  titlefont={'size': 24},
                  width=1000, height=1000,
                  xaxis_showgrid=False,
                  xaxis={'side': 'bottom'},
                  yaxis_showgrid=False,
                  yaxis_autorange='reversed',                   
                  paper_bgcolor=None,
                  )
fig.show()
```

```python
Vars = list(data)
cor = data.corr()

for var in Vars:
    cor_target = abs(cor[var])
    relevant_features = cor_target[cor_target > 0.5].index.tolist()
    print(f"Relevant features for '{var}': {relevant_features}")
```
